```python
# Calculate runs scored by each batter per match
top_batsmen_score = deliveries.groupby(['match_id', 'batter'])['batsman_runs'].sum().reset_index(name='Innings Runs')

# Calculate number of balls faced by each batter per match
batsman_ball_faced = deliveries.groupby(['match_id', 'batter'])['batsman_runs'].count().reset_index(name='Balls Faced')
```

```python
# Merge both DataFrames on match_id and batter
batsmen_performance = pd.merge(top_batsmen_score, batsman_ball_faced, how='inner', on=['match_id', 'batter'])

# Calculate strike rate
batsmen_performance['Strike Rate For Match'] = (batsmen_performance['Innings Runs'] / batsmen_performance['Balls Faced']) * 100

# Display results
batsmen_performance.head(10)
```

```python
balls_faced_all = deliveries.groupby('batter')['batsman_runs'].count().reset_index(name='Balls Faced')
balls_faced_all = balls_faced_all.sort_values(by='Balls Faced', ascending=False)

# Display top 10
print(balls_faced_all.head(10))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Top 10 innings by strike rate
top_strike_innings = batsmen_performance.sort_values(by='Strike Rate For Match', ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x='batter', y='Strike Rate For Match', data=top_strike_innings, palette='magma')
plt.title('Top 10 Highest Strike Rate Innings')
plt.xlabel('Batter')
plt.ylabel('Strike Rate')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Group by batter and average their strike rate
avg_strike_rate = batsmen_performance.groupby('batter')['Strike Rate For Match'].mean().reset_index()
avg_strike_rate = avg_strike_rate.sort_values(by='Strike Rate For Match', ascending=False).head(10)

# Plot
plt.figure(figsize=(12,6))
sns.barplot(x='batter', y='Strike Rate For Match', data=avg_strike_rate, palette='viridis')
plt.title('Top 10 Batters by Average Strike Rate')
plt.xlabel('Batter')
plt.ylabel('Average Strike Rate')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


match_bowler_runs = deliveries.groupby(['match_id', 'bowler'])['total_runs'].sum().reset_index(name='Runs Conceded')

wickets = deliveries[deliveries['dismissal_kind'].notna() & (deliveries['dismissal_kind'] != 'run out')]
match_bowler_wickets = wickets.groupby(['match_id', 'bowler'])['dismissal_kind'].count().reset_index(name='Wickets')


match_bowler_performance = pd.merge(match_bowler_runs, match_bowler_wickets, how='inner', on=['match_id', 'bowler'])


match_bowler_performance['Runs per Wicket'] = np.round(
    match_bowler_performance['Runs Conceded'] / match_bowler_performance['Wickets'], 2
)

# Step 4: Get top 10 performances with best Runs/Wicket ratio
top_bowler_efficiency = match_bowler_performance.sort_values(by='Runs per Wicket').head(10)

# Step 5: Plot line chart
plt.figure(figsize=(12,6))
sns.lineplot(data=top_bowler_efficiency, x='bowler', y='Runs per Wicket', marker='o', linewidth=2.5, color='darkblue')

plt.title('Top 10 Bowlers by Best Runs per Wicket Ratio', fontsize=14)
plt.xlabel('Bowler', fontsize=12)
plt.ylabel('Runs per Wicket', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
